```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/chayaphon/DADS5001/main/Assignment_3/data/2021-2022.csv')
display(df)
```

```python
# Slecting interesting column and check null data
cols = ['HomeTeam','AwayTeam','FTR']
df = df[cols]
df.info()
```

```python
# Calculating home, away, and total win rates for each team
home_wins = df[df['FTR'] == 'H'].groupby('HomeTeam').size()
away_wins = df[df['FTR'] == 'A'].groupby('AwayTeam').size()
total_matches_home = df.groupby('HomeTeam').size()
total_matches_away = df.groupby('AwayTeam').size()
total_wins = home_wins.add(away_wins, fill_value=0)
total_matches = total_matches_home.add(total_matches_away, fill_value=0)
home_win_rate = ((home_wins / total_matches_home) * 100).round(2)
away_win_rate = ((away_wins / total_matches_away) * 100).round(2)
win_rate = ((total_wins / total_matches) * 100).round(2)
df_win_rates = pd.DataFrame({
    'Home Win Rate %': home_win_rate,
    'Away Win Rate %': away_win_rate,
    'Total Win Rate %': win_rate
})
df_win_rates.sort_values(by='Total Win Rate %', ascending=False)
```

```python
# Defining the median of Home and Away Win Rates to split the quadrants
median_home_win_rate = df_win_rates['Home Win Rate %'].median()
median_away_win_rate = df_win_rates['Away Win Rate %'].median()
```

```python
# Data Visualization using Plotly

import plotly.express as px

# Create a scatter plot using Plotly Express
fig = px.scatter(df_win_rates, x='Away Win Rate %', y='Home Win Rate %', color=df_win_rates.index,
                 color_continuous_scale='tab20', text=df_win_rates.index)

# Add lines for median Home and Away Win Rates
fig.add_shape(type='line', x0=0, x1=100, y0=median_home_win_rate, y1=median_home_win_rate,
              line=dict(color='grey', dash='dash'))
fig.add_shape(type='line', x0=median_away_win_rate, x1=median_away_win_rate, y0=0, y1=100,
              line=dict(color='grey', dash='dash'))

# Update layout with title, axis labels, and limits
fig.update_layout(title='English Premier League 2021-2022 Win Rate',
                  title_x=0.5,
                  xaxis_title='Away Win Rate %', yaxis_title='Home Win Rate %',
                  xaxis_range=[0, 100], yaxis_range=[0, 100])

# Move legend outside the plot
fig.update_layout(legend=dict(title='Teams'), legend_x=1, legend_y=0.5)

# Adjust the size of the plot for better label visibility
fig.update_layout(width=1000, height=800) 
fig.update_traces(textposition='top center')
fig.update_traces(textfont=dict(color='rgba(0,0,0,0.5)')) 

# Show the plot
#fig.write_image("plotly_output.png")
fig.show()
```
